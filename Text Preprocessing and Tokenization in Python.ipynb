```python
# Code Cell 1  - Library and data import. Showing  summary of  input data.
# Importing necessary libaries
import pandas as pd
import re
from collections import Counter
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk import pos_tag, ne_chunk
from nltk.tree import Tree
from datetime import datetime
import nltk
nltk.download('stopwords')
nltk.download('punkt')  # For tokenization
nltk.download('averaged_perceptron_tagger')  # For POS tagging
nltk.download('maxent_ne_chunker')  # For named entity recognition
nltk.download('words')  # For NER processing

# Loading the dataset
file_path = 'C:\\Users\\balbi\\Downloads\\restaurant_reviews_az.csv' 
df = pd.read_csv(file_path)

# Displaying basic information about the dataset
print("Dataset Information:")
print(df.info())

# Displaying the summary statistics of the dataset
print("\nSummary Statistics:")
print(df.describe(include='all'))

# Showing the first few rows of the dataset
print("\nFirst 5 Rows of the Dataset:")
print(df.head())
```

```python
# Code Cell 2 - Selecting 1-star and 5-star reviews from dataset

# Select reviews with 1 star
one_star_reviews = df[df['stars'] == 1]['text'].tolist()

# Select reviews with 5 stars
five_star_reviews = df[df['stars'] == 5]['text'].tolist()

# Displaying  count of 1-star and 5-star reviews
print(f"Number of 1-star reviews: {len(one_star_reviews)}")
print(f"Number of 5-star reviews: {len(five_star_reviews)}")

# Displaying  sample of the reviews
print("\nSample 1-star review:")
print(one_star_reviews[0] if one_star_reviews else "No 1-star reviews found.")

print("\nSample 5-star review:")
print(five_star_reviews[0] if five_star_reviews else "No 5-star reviews found.")
```

```python
# Code Cell 3 - Apply necessary text processing techniques on selected review.
def preprocess_text(reviews):
    stop_words = set(stopwords.words('english'))  # Load stopwords
    processed_reviews = []

    for review in reviews:
        # Remove special characters, numbers, and punctuation
        review = re.sub(r'[^a-zA-Z\s]', '', review)

        # Convert text to lowercase
        review = review.lower()

        # Tokenize the text
        words = word_tokenize(review)

        # Remove stopwords
        words = [word for word in words if word not in stop_words]

        # Remove short words (length < 2)
        words = [word for word in words if len(word) > 1]

        processed_reviews.append(words)

    return processed_reviews

# Apply text preprocessing to 1-star and 5-star reviews
processed_one_star = preprocess_text(one_star_reviews)
processed_five_star = preprocess_text(five_star_reviews)

# Display the first processed review for each category
print("Sample processed 1-star review:")
print(processed_one_star[0] if processed_one_star else "No processed reviews found.")

print("\nSample processed 5-star review:")
print(processed_five_star[0] if processed_five_star else "No processed reviews found.")
```

```python
# Code Cell 4 - Finding the top 20 frequently used nouns in 1-star and 5-star reviews

def extract_top_nouns(processed_reviews):
    # Flatten the list of tokenized reviews
    all_words = [word for review in processed_reviews for word in review]
    
    # Perform POS tagging to identify nouns
    tagged_words = pos_tag(all_words)
    
    # Extract nouns (NN for singular noun, NNS for plural noun)
    nouns = [word for word, tag in tagged_words if tag in ('NN', 'NNS')]
    
    # Count the top 20 most common nouns
    return Counter(nouns).most_common(20)

# Getting top nouns for 1-star reviews
top_nouns_one_star = extract_top_nouns(processed_one_star)
print("Top 20 Nouns in 1-star Reviews:")
print(top_nouns_one_star)

# Get top nouns for 5-star reviews
top_nouns_five_star = extract_top_nouns(processed_five_star)
print("\nTop 20 Nouns in 5-star Reviews:")
print(top_nouns_five_star)
```

```python
# Code Cell 7 -  Top 20 frequently used named entities from the selected reviews.
# Ensure necessary NLTK resources are available
nltk.download('maxent_ne_chunker')
nltk.download('words')
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')

def extract_named_entities(reviews):
    all_entities = []
    for review in reviews:
        # Tokenize and POS tag the review
        words = word_tokenize(review)
        tagged_words = pos_tag(words)
        
        # Perform Named Entity Recognition (NER)
        chunked = ne_chunk(tagged_words)
        
        for chunk in chunked:
            if isinstance(chunk, Tree):  # Check if chunk is a named entity
                entity = " ".join(c[0] for c in chunk)  # Combine words forming the entity
                all_entities.append(entity)
    
    # Get the 20 most common named entities
    return Counter(all_entities).most_common(20)

# Extract named entities from both 1-star and 5-star reviews
named_entities_one_star = extract_named_entities(one_star_reviews)
named_entities_five_star = extract_named_entities(five_star_reviews)

# Display results
print("Top 20 Named Entities in 1-star Reviews:")
print(named_entities_one_star)

print("\nTop 20 Named Entities in 5-star Reviews:")
print(named_entities_five_star)
```
